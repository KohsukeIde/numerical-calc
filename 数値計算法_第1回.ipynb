# 課題4

```python
import numpy as np
import matplotlib.pyplot as plt

# xの値(-10から10まで）
x = np.arange(-10, 11, 0.1)

y = x**3 - 2*x**2 + 5*x + 8

plt.plot(x, y)

plt.title("Graph of f(x) = x^3 - 2x^2 + 5x + 8")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from CoolProp.CoolProp import PropsSI


temperature = 32.97  # Kelvin

# range of pressure (Pa)
pressure_min = 0.01 * 1e6  # convert MPa to Pa
pressure_max = 3 * 1e6  # convert MPa to Pa
pressures = np.linspace(pressure_min, pressure_max, 100)

# calculate densities for each pressure
densities = [PropsSI('D', 'P', p, 'T', temperature, 'Hydrogen') for p in pressures]

plt.figure(figsize=(8, 6))
plt.plot(pressures * 1e-6, densities, label='Hydrogen at 32.97 K')
plt.xlabel('Pressure (MPa)')
plt.ylabel('Density (kg/m³)')
plt.title('Density/Pressure correlation for Hydrogen(32.97 K)')
plt.legend()
plt.grid(True)
plt.show()
```
